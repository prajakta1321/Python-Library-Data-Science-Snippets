# Find the best fitting line in Numpy

```python
import numpy as np
```

```python
# first initialize 2 numpy arrays xs and ys with the first 5 positive integers
xs = np.array([1,2,3,4,5], dtype = np.float64)
ys = np.array([1,2,3,4,5], dtype = np.float64)

# using a function calculate the optimal values of m (slope) and b (y-intercept)
def best_fit_slope(xs, ys):
  m = ((np.mean(xs) * np.mean(ys)) - np.mean(xs * ys)) / ((np.mean(xs) * np.mean(xs)) - np.mean(xs * xs))
  b = np.mean(ys) - m * np.mean(xs)
  return m, b

m , b = best_fit_slope(xs, ys)

print('m:', m, 'b:', b)
```

# CALCULATING MSE BY SUCCESSIVE APPROXIMATION FOR A SET OF POINTS IN 2D PLANE

```python
import numpy as np
import matplotlib.pyplot as plt

# we first take 2 simple arrays
X = [0,0.12,0.25,0.27,0.38,0.42,0.44,0.55,0.92,1.0]
Y = [0,0.15,0.54,0.51,0.34,0.1, 0.19,0.53,1.0,0.58]
```

```python
# the costs array is initialized as an empty array, and we will append successive cost approximations to this array.
costs = []

# The variables W and b correspond to the slope and yintercept, and they are initialized with the foll. values.

#Step 1: Parameter initialization
W = 0.45 # the initial slope
b = 0.75 # the initial y-intercept
```

# Linear Regression

```python
# a for loop that executes 100 times.
# During each iteration, the variables Y_pred, loss_error, and cost are computed, and they correspond to the predicted value, the error, and the cost, respectively

for i in range(1, 100):
#Step 2: Calculate Cost
  Y_pred = np.multiply(W, X) + b         # predicted value
  loss_error = 0.5 * (Y_pred - Y)**2      # error
  cost = np.sum(loss_error)/10           # cost

  #Step 3: Calculate dw and db
  db = np.sum((Y_pred - Y))
  dw = np.dot((Y_pred - Y), X)
  costs.append(cost)

  #Step 4: Update parameters:
  W = W - 0.01*dw
  b = b - 0.01*db
  if i%10 == 0:
    print("Cost at", i,"iteration = ", cost)

#Step 5: Repeat via a for loop with 1000 iterations
```

```python
#Plot cost versus # of iterations
print("W = ", W,"& b = ", b)
plt.plot(costs)
plt.title("Cost Vs iterations")
plt.ylabel('cost')
plt.xlabel('iterations (per tens)')
plt.show()
```

# CALCULATING MSE BY SUCCESSIVE APPROXIMATION

```python
plt.plot(X,Y)
#plt.show()
```

```python
costs = []
#Step 1: Parameter initialization
W = 0.45
b = 0.75
epochs = 100
lr = 0.001
```

```python
for j in range(1, epochs):
  for i in range(1, 100):
#Step 2: Calculate Cost
    Y_pred = np.multiply(W, X) + b
    Loss_error = 0.5 * (Y_pred - Y)**2
    cost = np.sum(Loss_error)/10
#Step 3: Calculate dW and db
    db = np.sum((Y_pred - Y))
    dw = np.dot((Y_pred - Y), X)
    costs.append(cost)
#Step 4: Update parameters:
    W = W - lr*dw
    b = b - lr*db
    if i%50 == 0:
      print("Cost at epoch", j,"= ", cost)
```

```python
#Plot cost versus # of iterations
print("W = ", W,"& b = ", b)
plt.plot(costs)
plt.ylabel('cost')
plt.xlabel('iterations (per tens)')
plt.show()
```
